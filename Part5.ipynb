```python
import seaborn as sns

def visualize_data(cosine_matrix, texts_processed):

    # Відображення теплової карти матриці косинусної подібності
    plt.figure(figsize=(10, 8))
    sns.heatmap(cosine_matrix, annot=True, cmap="Blues",
                xticklabels=[f"Doc {i+1}" for i in range(len(texts_processed))],
                yticklabels=[f"Doc {i+1}" for i in range(len(texts_processed))])
    plt.title('Cosine Similarity Matrix')
    plt.show()

visualize_data(cosine_matrix, texts_processed)
```
