```python
import random
import string

def get_texts(num_texts=20, text_length=40):

    texts = []
    for _ in range(num_texts):
        # Генеруємо випадковий текст
        text = ''.join(random.choices(string.ascii_lowercase + " ", k=text_length))
        texts.append(text)
    return texts

# Завантажуємо тексти (тут згенеруємо випадкові дані для прикладу)
texts = get_texts(num_texts=20, text_length=40)

# Обробка текстів
texts_processed = process_texts(texts)

# Візуалізація хмари слів для перевірки результату обробки
visualize_texts(texts_processed)

# Обчислення TF-IDF та косинусної подібності
cosine_matrix = calculate_tfidf(texts_processed)

# Виведення матриці косинусної подібності (можна вивести частково)
print(cosine_matrix[:5, :5])  # Перевірка перших п'яти значень для візуалізації

# Візуалізація косинусної подібності для всіх текстів
visualize_data(cosine_matrix, texts_processed)
```
